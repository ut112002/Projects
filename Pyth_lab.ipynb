```python
class Car:
    def __init__(self, model_name, color, price, top_speed):
        self.model_name = model_name
        self.color = color
        self.price = price
        self.top_speed = top_speed

    def display_details(self):
        print("Car Details:")
        print(f"Model Name: {self.model_name}")
        print(f"Color: {self.color}")
        print(f"Price: ${self.price}")
        print(f"Top Speed: {self.top_speed} km/h")
        print()

# Example usage with constructor
car1 = Car("Toyota Camry", "Blue", 25000, 180)
car1.display_details()

# Example usage without constructor
car2 = Car(None, None, None, None)
car2.model_name = "Honda Accord"
car2.color = "Silver"
car2.price = 28000
car2.top_speed = 170
car2.display_details()
##########################################
class Bank:
    def __init__(self, bank_name, number_cust):
        self.bank_name = bank_name
        self.number_cust = number_cust

    def display(self):
        print("Bank Details:")
        print(f"Bank Name: {self.bank_name}")
        print(f"Number of Customers: {self.number_cust}")
        print()


class GovtBank(Bank):
    def __init__(self, bank_name, number_cust, branch_name, ifsc_code):
        super().__init__(bank_name, number_cust)
        self.branch_name = branch_name
        self.ifsc_code = ifsc_code

    def display(self):
        super().display()
        print("Govt Bank Details:")
        print(f"Branch Name: {self.branch_name}")
        print(f"IFSC Code: {self.ifsc_code}")
        print()


class PrivateBank(Bank):
    def __init__(self, bank_name, number_cust, branch_name, ifsc_code):
        super().__init__(bank_name, number_cust)
        self.branch_name = branch_name
        self.ifsc_code = ifsc_code

    def display(self):
        super().display()
        print("Private Bank Details:")
        print(f"Branch Name: {self.branch_name}")
        print(f"IFSC Code: {self.ifsc_code}")
        print()


# Example usage
govt_bank = GovtBank("Public Bank", 1000, "Main Branch", "GOV12345")
private_bank = PrivateBank("Private Bank Ltd.", 500, "Downtown Branch", "PVT67890")

govt_bank.display()
private_bank.display()
```

```python
class Time:
    def __init__(self, hour=0, minute=0, second=0):
        self.hour = hour
        self.minute = minute
        self.second = second

    def display(self):
        print(self.hour,":",self.minute,":",self.second)

    def __add__(self, other):
        total_seconds_self = self.hour * 3600 + self.minute * 60 + self.second
        total_seconds_other = other.hour * 3600 + other.minute * 60 + other.second
        total_seconds_sum = total_seconds_self + total_seconds_other

        hours, remainder = divmod(total_seconds_sum, 3600)
        minutes, seconds = divmod(remainder, 60)

        return Time(hours, minutes, seconds)

    def __sub__(self, other):
        total_seconds_self = self.hour * 3600 + self.minute * 60 + self.second
        total_seconds_other = other.hour * 3600 + other.minute * 60 + other.second
        total_seconds_diff = abs(total_seconds_self - total_seconds_other)

        hours, remainder = divmod(total_seconds_diff, 3600)
        minutes, seconds = divmod(remainder, 60)

        return Time(hours, minutes, seconds)

# Example usage
time1 = Time(3, 45, 30)
time2 = Time(1, 30, 15)

print("Time 1:")
time1.display()

print("\nTime 2:")
time2.display()

print("\nAddition of Time 1 and Time 2:")
result_addition = time1 + time2
result_addition.display()

print("\nSubtraction of Time 1 and Time 2:")
result_subtraction = time1 - time2
result_subtraction.display()
```

```python
def count_sentences_words_characters(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()

            sentences = content.split('.') + content.split('?')
            num_sentences = len(sentences)

            # Count words by splitting on whitespace
            words = content.split()
            num_words = len(words)

            # Count characters
            num_characters = len(content)

            # Display the results
            print(f"Number of Sentences: {num_sentences}")
            print(f"Number of Words: {num_words}")
            print(f"Number of Characters: {num_characters}")

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Example usage
filename = input("Enter the filename: ")
count_sentences_words_characters(filename)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from sklearn.datasets import load_iris

# Load the Iris dataset from scikit-learn
iris = load_iris()
iris_data = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

# i. Scatter plot
plt.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'], c=iris_data['target'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.show()

# ii. Calculate mean
mean_sepal_length = np.mean(iris_data['sepal length (cm)'])
mean_sepal_width = np.mean(iris_data['sepal width (cm)'])
print(f'Mean Sepal Length: {mean_sepal_length:.2f} cm')
print(f'Mean Sepal Width: {mean_sepal_width:.2f} cm')

# iii. Calculate median
median_sepal_length = np.median(iris_data['sepal length (cm)'])
median_sepal_width = np.median(iris_data['sepal width (cm)'])
print(f'Median Sepal Length: {median_sepal_length:.2f} cm')
print(f'Median Sepal Width: {median_sepal_width:.2f} cm')

# iv. Calculate standard deviation
std_sepal_length = np.std(iris_data['sepal length (cm)'])
std_sepal_width = np.std(iris_data['sepal width (cm)'])
print(f'Standard Deviation Sepal Length: {std_sepal_length:.2f} cm')
print(f'Standard Deviation Sepal Width: {std_sepal_width:.2f} cm')

# v. Calculate variance
var_sepal_length = np.var(iris_data['sepal length (cm)'])
var_sepal_width = np.var(iris_data['sepal width (cm)'])
print(f'Variance Sepal Length: {var_sepal_length:.2f} cm^2')
print(f'Variance Sepal Width: {var_sepal_width:.2f} cm^2')

# vi. Calculate slope between points
slope, intercept, r_value, p_value, std_err = linregress(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'])
print(f'Slope between Sepal Length and Sepal Width: {slope:.2f}')

# vii. Draw regression line
plt.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'], c=iris_data['target'], cmap='viridis')
plt.plot(iris_data['sepal length (cm)'], intercept + slope * iris_data['sepal length (cm)'], 'r', label='Regression Line')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Regression Line between Sepal Length and Sepal Width')
plt.legend()
plt.show()
```

```python
def sieve_of_eratosthenes(limit):
    primes = [True] * (limit + 1)
    primes[0] = primes[1] = False

    for num in range(2, int(limit**0.5) + 1):
        if primes[num]:
            for multiple in range(num * num, limit + 1, num):
                primes[multiple] = False

    return sum(x for x in range(2, limit + 1) if primes[x])
```
